```python
def merge_dicts(*dicts):
    result_dict = {'data': []}
    specialisms = {}

    for d in dicts:
        for key, value in d.items():
            if key == 'CaseReference':
                result_dict[key] = value
            elif key == 'Specialisms':
                if value not in specialisms:
                    specialisms[value] = {'ID': value, 'Name': None}

                current_specialism = specialisms[value]

                if key in result_dict:
                    result_dict[key]['Specialisms'].append(current_specialism)
                else:
                    result_dict[key] = {'Specialisms': [current_specialism]}
            elif key == 'Name':
                current_specialism['Name'] = value
            else:
                result_dict[key] = value

    return result_dict

# Example usage with two dictionaries
dict1 = {'CaseReference': 123, 'SpecialismID': 456, 'Name': 'Specialism1'}
dict2 = {'CaseReference': 123, 'SpecialismID': 765, 'Name': 'Specialism2'}

result_dict = merge_dicts(dict1, dict2)
print(result_dict)
```
